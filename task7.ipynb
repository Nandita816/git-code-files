```python
# Creating a list
fruits = ["Apple", "Banana", "Orange"]

print("Original List:", fruits)

# append() - Adds an item at the end
fruits.append("Mango")
print("After append():", fruits)

# insert() - Inserts an item at a specified position
fruits.insert(1, "Grapes")
print("After insert():", fruits)

# remove() - Removes a specified item
fruits.remove("Banana")
print("After remove():", fruits)

# pop() - Removes item from specified index (default: last item)
fruits.pop()
print("After pop():", fruits)

# sort() - Sorts the list in ascending order
fruits.sort()
print("After sort():", fruits)

# reverse() - Reverses the list order
fruits.reverse()
print("After reverse():", fruits)

# count() - Counts occurrences of an item
print("Count of Apple:", fruits.count("Apple"))

# index() - Returns index of an item
print("Index of Orange:", fruits.index("Orange"))

# len() - Returns total number of items
print("Length of List:", len(fruits))
```
